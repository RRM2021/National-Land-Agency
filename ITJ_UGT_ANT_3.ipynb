Se crea un método para filtrar columnas asociadas a ITJ

```python
import pandas as pd

# Assuming df is your DataFrame
# Iterate over the column names and select only the columns that contain 'ITJ' as a substring
filtered_columns = [col for col in df.columns if 'ITJ' in col]

# Create a new DataFrame with the selected columns
filtered_df = df[filtered_columns]
print(filtered_columns)
```

```python
filtered_df
```

```python
import pandas as pd


# Display original column names
print("Original column names:")
print(filtered_df.columns)

# Trim the specific word combination from column names
filtered2_df = filtered_df.rename(columns=lambda x: x.replace('ITJ', ''))

# Display updated column names
print("\nUpdated column names:")
print(filtered2_df.columns)
```

Si se van a producir los ITJs usando herramientas de Python, se crea un directorio, se crean archivos de texto con los campos del dataframe, y se almacenan en el directorio.

```python
import pandas as pd
import os

# Create a directory to save the documents if it doesn't exist
output_dir = 'itjs'
os.makedirs(output_dir, exist_ok=True)

# Iterate over each row
for idx, row in filtered2_df.iterrows():
    # Create a string with the row's values concatenated with line breaks
    document_content = "\n".join([f"{col}: {value}" for col, value in row.items()])

    # Define the file name (e.g., using the row index)
    file_name = os.path.join(output_dir, f"document_{idx}.txt")

    # Write the string to a text file
    with open(file_name, 'w') as file:
        file.write(document_content)

print("Documents created successfully.")
```

Se identifica la ruta donde se creó el directorio con los archivos de texto

```python
import os
from docx import Document
from google.colab import files

# Directory and file paths
directory_path = '/content/itjs_3'
text_file_name = 'document_34.txt'
text_file_path = os.path.join(directory_path, text_file_name)
word_file_name = 'document_34.docx'
word_file_path = os.path.join(directory_path, word_file_name)

# Check if the text file exists
if os.path.exists(text_file_path):
    # Read the content of the text file
    with open(text_file_path, 'r') as file:
        text_content = file.read()

    # Create a new Word document
    doc = Document()

    # Add the text content to the Word document
    for line in text_content.split('\n'):
        doc.add_paragraph(line)

    # Save the Word document
    doc.save(word_file_path)
    print(f"Word document created at {word_file_path}")

    # Download the Word document
    files.download(word_file_path)
else:
    print(f"Text file not found: {text_file_path}")
```

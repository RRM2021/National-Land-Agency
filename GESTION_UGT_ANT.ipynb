```python
import pandas as pd
from pyxlsb import open_workbook as open_xlsb

# Replace 'path_to_file.xlsb' with the actual path to your uploaded XLSB file
df = []
with open_xlsb('/content/MATRIZ_DE_GESTION_UGT-ANTIOQUIA-2024 - CORTE DEL 3 DE ABRIL.xlsb') as wb:
    with wb.get_sheet(1) as sheet:
        for row in sheet.rows():
            df.append([item.v for item in row])

df = pd.DataFrame(df[1:], columns=df[0])
```

*La matriz consta de 3414 filas y 77 columnas*

```python
# Initialize dictionaries to store counts
null_counts = {}
not_null_counts = {}

# Iterate over columns
for col in df.columns:
    null_counts[col] = df[col].isnull().sum()
    not_null_counts[col] = df[col].notnull().sum()

# Create DataFrame from dictionaries
null_counts_df = pd.DataFrame.from_dict(null_counts, orient='index', columns=['Null Count'])
not_null_counts_df = pd.DataFrame.from_dict(not_null_counts, orient='index', columns=['Not Null Count'])

# Concatenate DataFrames
result_df = pd.concat([null_counts_df, not_null_counts_df], axis=1)

print("Null and Not Null Value Counts:")
print(result_df)
result_df
```

```python
import pandas as pd

# Initialize an empty dictionary to store value-wise duplicate counts
value_duplicate_counts = {}

# Iterate over each column
for col in df.columns:
    # Count the duplicates for the current column (excluding null values)
    duplicate_count = len(df.dropna(subset=[col])) - len(df.drop_duplicates(subset=col, keep=False))
    value_duplicate_counts[col] = duplicate_count

# Create a DataFrame from the dictionary
result_df2 = pd.DataFrame(value_duplicate_counts.items(), columns=['Column', 'Duplicate Count'])

print("Value-wise duplicate counts for each column (excluding null values):")
print(result_df2)
result_df2
```
